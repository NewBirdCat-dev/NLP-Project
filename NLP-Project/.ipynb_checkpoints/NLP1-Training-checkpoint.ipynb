```python
# ==========================================================
# RESUME SCREENING PIPELINE (COLAB READY)
# Supports: TF-IDF, Skills, GloVe + Grid Search + Metrics + Skill Evaluation + Visuals
# ==========================================================

import os, re, json, joblib, warnings              # Import OS utilities, regex, JSON, model saving, warnings
import numpy as np                                 # Numerical computing
import pandas as pd                                # DataFrames
from scipy.sparse import hstack, csr_matrix        # Sparse matrix utilities for combining features
import matplotlib.pyplot as plt                    # Plotting
import seaborn as sns                              # Statistical visualizations

from sklearn.feature_extraction.text import TfidfVectorizer  # TF-IDF vectorizer
from sklearn.model_selection import train_test_split, GridSearchCV  # Train/test split & hyperparameter search
from sklearn.linear_model import LogisticRegression              # Logistic regression classifier
from sklearn.ensemble import RandomForestClassifier             # Random forest classifier
from sklearn.naive_bayes import MultinomialNB                   # Naive Bayes classifier
from sklearn.metrics import f1_score, classification_report, accuracy_score, precision_score, recall_score, confusion_matrix  # Metrics

from wordcloud import WordCloud                                # Word cloud visualization
from rapidfuzz import fuzz                                     # Fuzzy matching for skill extraction
import nltk                                                    # NLP utilities
from nltk.corpus import stopwords                               # English stopwords
from nltk.stem import WordNetLemmatizer                         # Lemmatizer

warnings.filterwarnings("ignore")                               # Disable warnings for clean output

# ---------------------- CONFIG ----------------------
DATA_PATH = "UpdatedResumeDataSet.csv"                          # Path to dataset CSV
MODEL_SAVE_PATH = "./best_resume_model.joblib"                  # Path to save best ML model
TFIDF_SAVE_PATH = "./tfidf_vectorizer.joblib"                   # Path to save TF-IDF vectorizer
SKILL_VOCAB_SAVE_PATH = "./skill_vocab.json"                    # Path to save skill vocabulary
GLOVE_PATH = "./glove.6B/glove.6B.300d.txt"                     # Path to GloVe embeddings file

# ---------------------- SKILL LIST ----------------------
SKILLS = [                                                      # Predefined list of technical & soft skills
    'python','java','javascript','typescript','c','c++','c#','r','swift','kotlin','go','scala','php',
    'html','css','sql','bash','shell','powershell','dart','machine learning','deep learning','nlp',
    'natural language processing','computer vision','data science','data analysis','data visualization',
    'predictive modeling','feature engineering','tensorflow','keras','pytorch','scikit-learn','pandas',
    'numpy','matplotlib','seaborn','plotly','xgboost','lightgbm','catboost','statsmodels','spacy','nltk',
    'huggingface','transformers','opencv','streamlit','aws','azure','gcp','google cloud','docker','kubernetes',
    'jenkins','terraform','ansible','git','github','gitlab','ci/cd','linux','unix','mysql','postgresql',
    'sqlite','mongodb','snowflake','bigquery','oracle','elasticsearch','hadoop','spark','airflow','databricks',
    'excel','tableau','powerbi','looker','metabase','dash','react','angular','vue','nodejs','django','flask',
    'fastapi','spring boot','api development','microservices','communication','teamwork','leadership',
    'problem solving','critical thinking','adaptability','collaboration','creativity','attention to detail'
]
SKILL_DICT = set([s.lower() for s in SKILLS])                    # Convert skills to lowercase for matching

# ---------------------- NLTK SETUP ----------------------
try:
    STOPWORDS = set(stopwords.words("english"))                  # Load English stopwords
except:
    nltk.download("stopwords")                                   # Download stopwords if missing
    STOPWORDS = set(stopwords.words("english"))

nltk.download("punkt")                                           # Download tokenizer
nltk.download("wordnet")                                         # Download lemmatizer dictionary
LEMMATIZER = WordNetLemmatizer()                                 # Instantiate lemmatizer

# ---------------------- TEXT PREPROCESSING ----------------------
def clean_text(text):
    if not isinstance(text, str): return ""                      # Handle non-string inputs
    text = re.sub(r'<[^>]+>', ' ', text)                         # Remove HTML tags
    text = re.sub(r'http\S+|www\.\S+', ' ', text)                # Remove URLs
    text = re.sub(r'\S+@\S+', ' ', text)                         # Remove emails
    text = re.sub(r'[^A-Za-z0-9 ]', ' ', text)                   # Keep only alphanumeric characters
    text = re.sub(r'\s+', ' ', text).strip()                     # Normalize whitespace
    return text.lower()                                          # Lowercase output

def preprocess_text(text):
    tokens = [LEMMATIZER.lemmatize(w) for w in clean_text(text).split() if w not in STOPWORDS]  
    return " ".join(tokens)                                      # Rejoin into cleaned text string

# ---------------------- SKILL EXTRACTION ----------------------
def extract_skills(text):
    text_low = text.lower()                                      # Lowercase for matching
    return {s for s in SKILL_DICT if s in text_low}              # Return detected skills

def skills_to_vector(skills, vocab):
    vec = np.zeros(len(vocab), dtype=int)                        # Initialize skill vector
    idx = {s:i for i,s in enumerate(vocab)}                      # Map skills to positions
    for s in skills:
        if s in idx: vec[idx[s]] = 1                             # Mark present skills as 1
    return vec

# ---------------------- SKILL EXTRACTION EVALUATION ----------------------
def normalize_text(text):
    text = text.lower()                                          # Lowercase text
    typo_dict = {                                                # Dictionary of common typos
        "pyhton": "python",
        "jva": "java",
        "tensoflow": "tensorflow",
        "kubrnets": "kubernetes",
        "cv": "computer vision",
        "ml": "machine learning",
        "nlp": "natural language processing"
    }
    for k, v in typo_dict.items():
        text = text.replace(k, v)                                # Correct typos
    return text

def fuzzy_extract_skills_ngram(text, threshold=90):
    text = normalize_text(text)                                  # Normalize typos
    words = text.split()                                         # Tokenize
    ngrams = [' '.join(words[i:j]) for i in range(len(words))    # Create n-grams up to length 3
             for j in range(i+1, min(i+4, len(words)+1))]
    found = set()                                                # Store matched skills
    for skill in SKILL_DICT:                                     # Compare each skill
        if any(fuzz.ratio(skill, n) >= threshold for n in ngrams):  # Fuzzy match
            found.add(skill)
    return found

def evaluate_skill_extraction(test_texts, threshold=90):
    precisions, recalls = [], []                                 # Collect metrics
    for text in test_texts:
        extracted = fuzzy_extract_skills_ngram(text, threshold)  # Extract skills from text
        gt = {s.lower() for s in SKILLS if s in normalize_text(text)}  # Ground truth skill presence
        tp = len(extracted & gt)                                 # True positives
        precision = tp / len(extracted) if extracted else 0       # Precision score
        recall = tp / len(gt) if gt else 0                        # Recall score
        precisions.append(precision)                             # Append precision
        recalls.append(recall)                                   # Append recall
    print(f"ðŸ”¹ Average Skill Extraction Precision: {np.mean(precisions):.2f}")  
    print(f"ðŸ”¹ Average Skill Extraction Recall: {np.mean(recalls):.2f}")

    plt.figure(figsize=(8,5))                                    # Plot precision & recall
    plt.bar(["Precision","Recall"], [np.mean(precisions), np.mean(recalls)], color=['skyblue','salmon'])
    plt.title("Skill Extraction Performance")
    plt.ylim(0,1)
    plt.show()

    return precisions, recalls                                   # Return metric arrays

# ---------------------- GLOVE ----------------------
def download_glove():
    if not os.path.exists("./glove.6B/glove.6B.300d.txt"):        # Check if GloVe exists
        print("Downloading GloVe embeddings...")
        !wget -q http://nlp.stanford.edu/data/glove.6B.zip        # Download GloVe ZIP
        !unzip -q glove.6B.zip -d glove.6B                        # Extract zip
        !rm glove.6B.zip                                          # Remove ZIP
        print("âœ… GloVe downloaded to folder: /glove.6B/")

def load_glove():
    download_glove()                                              # Ensure GloVe exists
    glove = {}                                                    # Dictionary for embeddings
    with open(GLOVE_PATH, "r", encoding="utf8") as f:            # Open GloVe file
        for line in f:                                           # Read embeddings line-by-line
            vals = line.split()                                  # Split word + vector
            word = vals[0]
            vec = np.array(vals[1:], dtype=float)                # Convert to float vector
            glove[word] = vec                                     # Store embedding
    return glove

def glove_vector(text, glove, dim=300):
    vecs = [glove[w] for w in text.split() if w in glove]         # Collect word embeddings
    return np.mean(vecs, axis=0) if vecs else np.zeros(dim)       # Return mean or zeros

# ---------------------- DATASET ----------------------
def load_dataset(path):
    df = pd.read_csv(path)                                        # Load CSV file
    text_col = [c for c in df.columns if 'resume' in c.lower() or 'text' in c.lower()][0]  # Auto-detect resume column
    label_col = 'Category' if 'Category' in df.columns else df.columns[-1]  # Auto-detect label column
    df = df.dropna(subset=[text_col]).drop_duplicates(subset=[text_col])   # Remove empty or duplicate resumes
    df['clean_text'] = df[text_col].apply(clean_text)             # Apply cleaning
    df['preproc'] = df['clean_text'].apply(preprocess_text)       # Fully preprocess
    return df, text_col, label_col

# ---------------------- FEATURE BUILDING ----------------------
def build_features(df, tfidf=None, skill_vocab=None, glove=None):
    if 'skills' not in df.columns:
        df['skills'] = df['clean_text'].apply(extract_skills)     # Extract skills if missing
    if skill_vocab is None:
        skill_vocab = sorted({s for x in df['skills'] for s in x})  # Build skill vocabulary
    if tfidf is None:                                             # Fit TF-IDF if not provided
        tfidf = TfidfVectorizer(ngram_range=(1,3), max_features=8000)
        X_text = tfidf.fit_transform(df['preproc'])
    else:
        X_text = tfidf.transform(df['preproc'])                   # Transform using existing TF-IDF
    X_skill = csr_matrix(np.vstack([skills_to_vector(s, skill_vocab) for s in df['skills']]))  # Build skill vectors
    if glove:                                                     # If using GloVe
        X_glove = csr_matrix(np.vstack([glove_vector(t, glove) for t in df['clean_text']] ))    # Build GloVe features
        X_full = hstack([X_text, X_skill, X_glove])               # Hybrid feature stack
    else:
        X_full = hstack([X_text, X_skill])                        # TF-IDF + Skill only
    return X_text, X_full, tfidf, skill_vocab
# ---------------------- VISUALIZATION ----------------------
def visualize_text(df):
    before_text = " ".join(df['clean_text'])                      # Combine all cleaned text into a single string for word cloud
    wc_before = WordCloud(width=1000, height=600).generate(before_text)  # Generate word cloud for raw text

    after_text = " ".join(df['preproc'])                          # Combine preprocessed text into a single string
    wc_after = WordCloud(width=1000, height=600).generate(after_text)   # Generate word cloud for preprocessed text

    plt.figure(figsize=(14,6))                                    # Set figure size for side-by-side visualization
    plt.subplot(1,2,1)                                            # First subplot (before preprocessing)
    plt.title("Before Preprocessing")                              # Set title
    plt.imshow(wc_before)                                         # Display word cloud image
    plt.axis("off")                                                # Hide axes

    plt.subplot(1,2,2)                                            # Second subplot (after preprocessing)
    plt.title("After Preprocessing")                               # Set title
    plt.imshow(wc_after)                                          # Display word cloud image
    plt.axis("off")                                                # Hide axes
    plt.show()                                                     # Render both subplots

def plot_category_distribution(df, label_col):
    counts = df[label_col].value_counts()                         # Count number of resumes per category
    plt.figure(figsize=(10,6))                                    # Set figure size
    sns.barplot(x=counts.index, y=counts.values, palette="viridis")  # Bar plot showing category distribution
    plt.title("Job Category Distribution")                         # Set title
    plt.ylabel("Number of Resumes")                                # Set y-axis label
    plt.xticks(rotation=45)                                        # Rotate x-axis labels for readability
    plt.show()                                                     # Display plot

def plot_skill_frequency(df, top_n=20):
    all_skills = [s for skills in df['skills'] for s in skills]   # Flatten all skills from all resumes into a single list
    freq = pd.Series(all_skills).value_counts().head(top_n)       # Count occurrences and take top N skills
    plt.figure(figsize=(10,6))                                     # Set figure size
    sns.barplot(x=freq.values, y=freq.index, palette="magma")     # Horizontal bar plot of skill frequency
    plt.title(f"Top {top_n} Skills in Dataset")                   # Set title
    plt.xlabel("Frequency")                                       # Set x-axis label
    plt.show()                                                     # Display plot

# ---------------------- MODEL EVALUATION ----------------------
def evaluate_models(X_train_tfidf, X_train_full, X_test_tfidf, X_test_full, y_train, y_test):
    results = {}                                                  # Dictionary to store F1 scores of different models
    print("ðŸ” Logistic Regression + TF-IDF + Skills + GloVe")
    lr = GridSearchCV(LogisticRegression(max_iter=4000), {"C":[0.1,1,10]}, cv=3, scoring="f1_weighted")  # Grid search LR hyperparameters
    lr.fit(X_train_full, y_train)                                 # Train logistic regression on full feature set
    preds_lr = lr.predict(X_test_full)                            # Make predictions on test set
    results['Logistic Regression'] = f1_score(y_test, preds_lr, average="weighted")  # Calculate weighted F1

    print("ðŸŒ² Random Forest + TF-IDF + Skills + GloVe")
    rf = GridSearchCV(RandomForestClassifier(), {"n_estimators":[200,500],"max_depth":[None,20]}, cv=3, scoring="f1_weighted")  # Grid search RF
    rf.fit(X_train_full, y_train)                                 # Train RF on full feature set
    preds_rf = rf.predict(X_test_full)                             # Predictions
    results['Random Forest (Hybrid)'] = f1_score(y_test, preds_rf, average="weighted")  # Weighted F1

    print("ðŸ“¦ Naive Bayes (TF-IDF only)")
    nb = GridSearchCV(MultinomialNB(), {"alpha":[0.1,0.5,1.0]}, cv=3, scoring="f1_weighted")  # Grid search NB
    nb.fit(X_train_tfidf, y_train)                                 # Train NB on TF-IDF features only
    preds_nb = nb.predict(X_test_tfidf)                             # Predictions
    results['Naive Bayes'] = f1_score(y_test, preds_nb, average="weighted")  # Weighted F1

    print("ðŸŒ² Random Forest (TF-IDF + Skills only â€” Legacy)")
    rf_legacy = RandomForestClassifier()                            # Legacy RF classifier (no GloVe)
    rf_legacy.fit(X_train_tfidf, y_train)                           # Train on TF-IDF only
    preds_rf_legacy = rf_legacy.predict(X_test_tfidf)               # Predictions
    results['Random Forest (Legacy)'] = f1_score(y_test, preds_rf_legacy, average="weighted")  # Weighted F1

    print("\nâœ… MODEL PERFORMANCE (F1 weighted):")
    for k,v in results.items():                                     # Print all F1 scores
        print(f"{k}: {v:.4f}")

    best_model_name = max(results, key=results.get)                 # Find model with highest F1 score
    print(f"\nðŸ† BEST MODEL: {best_model_name}")

    best_model = {                                                  # Map model name to trained object
        'Logistic Regression': lr.best_estimator_,
        'Random Forest (Hybrid)': rf.best_estimator_,
        'Naive Bayes': nb.best_estimator_,
        'Random Forest (Legacy)': rf_legacy
    }[best_model_name]

    return best_model                                               # Return the best performing model

def model_evaluation_report(model, X_test, y_test, classes=None):
    preds = model.predict(X_test)                                   # Make predictions with selected model
    print("ðŸ“Š Classification Report:")
    print(classification_report(y_test, preds))                     # Print precision, recall, f1 per class
    acc = accuracy_score(y_test, preds)                              # Overall accuracy
    f1 = f1_score(y_test, preds, average="weighted")                 # Weighted F1 score
    prec = precision_score(y_test, preds, average="weighted")        # Weighted precision
    rec = recall_score(y_test, preds, average="weighted")            # Weighted recall
    print(f"Accuracy: {acc:.4f}, Weighted F1: {f1:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}")  # Print metrics

    cm = confusion_matrix(y_test, preds)                              # Confusion matrix
    plt.figure(figsize=(10,7))                                        # Set figure size
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=classes, yticklabels=classes)  # Plot heatmap
    plt.ylabel('Actual')                                              # y-axis label
    plt.xlabel('Predicted')                                           # x-axis label
    plt.title('Confusion Matrix Heatmap')                             # Set title
    plt.show()                                                        # Display heatmap

# ---------------------- FULL PIPELINE ----------------------
def full_pipeline():
    print("ðŸ“Œ Loading Dataset...")
    df, text_col, label_col = load_dataset(DATA_PATH)                # Load CSV dataset
    print(f"âœ… Dataset Loaded: {df.shape}")                           # Print shape
    print(df.head())                                                  # Show first 5 rows

    df['skills'] = df['clean_text'].apply(extract_skills)            # Extract skills from raw text

    visualize_text(df)                                               # Display before/after word clouds
    plot_category_distribution(df, label_col)                        # Plot job category distribution
    plot_skill_frequency(df)                                         # Plot top skills frequency

    print("â³ Loading GloVe embeddings... (can take 1â€“2 mins)")
    glove = load_glove()                                             # Load GloVe embeddings
    print("âœ… Loaded GloVe Embeddings")

    X_tfidf, X_full, tfidf, skill_vocab = build_features(df, glove=glove)  # Build TF-IDF, skill vectors, GloVe features
    y = df[label_col]                                                # Extract labels

    X_tr_tfidf, X_ts_tfidf, X_tr_full, X_ts_full, y_tr, y_ts = train_test_split(
        X_tfidf, X_full, y, test_size=0.2, stratify=y, random_state=42
    )                                                                # Split dataset into stratified train/test

    best_model = evaluate_models(X_tr_tfidf, X_tr_full, X_ts_tfidf, X_ts_full, y_tr, y_ts)  # Train and select best model

    if isinstance(best_model, MultinomialNB):                        # Select correct test features for evaluation
        X_eval = X_ts_tfidf
    elif isinstance(best_model, RandomForestClassifier) and best_model.n_features_in_ == X_ts_tfidf.shape[1]:
        X_eval = X_ts_tfidf
    else:
        X_eval = X_ts_full

    model_evaluation_report(best_model, X_eval, y_ts, classes=sorted(y_ts.unique()))  # Generate evaluation report

    TEST_RESUMES = [                                                 # Example resumes for skill extraction evaluation
        "Proficient in Pyhton, Jva, and ml frameworks like TEnsorflow and pytorch.",
        "Experience with NLP, CV, data analysis, and cloud(AWS, GCP).",
        "Strong in JavaScript, React, NodeJS, Docker, kubrnets and CI/CD pipelines."
    ]
    print("\nðŸ“Œ Evaluating Skill Extraction on Test Resumes...")
    evaluate_skill_extraction(TEST_RESUMES, threshold=90)            # Evaluate fuzzy skill extraction

    joblib.dump(best_model, MODEL_SAVE_PATH)                         # Save best model to disk
    joblib.dump(tfidf, TFIDF_SAVE_PATH)                              # Save TF-IDF vectorizer
    with open(SKILL_VOCAB_SAVE_PATH, "w") as f:
        json.dump(skill_vocab, f)                                    # Save skill vocabulary

    print("\nðŸ’¾ Model, vectorizer & skill vocab saved!")             # Confirmation message
    return best_model, tfidf, skill_vocab                             # Return trained model, vectorizer, vocab

# ---------------------- RUN PIPELINE ----------------------
best_model, tfidf, skill_vocab = full_pipeline()                     # Execute full pipeline and get trained artifacts
```
